```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

img = cv2.imread('./images/1_1.jpg')
```

```python
img.shape
```

```python
imgray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, imthres = cv2.threshold(imgray, 127, 255, cv2.THRESH_BINARY_INV)
```

```python
plt.imshow(imthres)
```

```python
contours, hierarchy = cv2.findContours(imgray, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)[-2:]
```

```python
len(contours)
```

```python
hierarchy
```

```python
external_contours = np.zeros(img.shape)
```

```python
cv2.drawContours(img, contours, -1, (0,255,0), 4)
```

```python
for i in contours:
    for j in i:
        cv2.circle(img, tuple(j[0]), 1, (255, 0, 0), -1)
```
